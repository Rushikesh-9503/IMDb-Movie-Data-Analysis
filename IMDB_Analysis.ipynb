#### Classify movies based on Ratings [Excellent,Good and Average]

```python
df.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating >=6.0:
        return "Good"
    else:
        return "Average"
```

#### Find unique values from Genre 

```python
df.columns
```

```python
df['Genre']
```

```python
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
```

```python
# we got this 2-D list
list1
```

```python
# converting 2-D list into 1-D
one_D = []
for item in list1:
    for item1 in item:
        one_D.append(item1)
```

```python
one_D
```

```python
# Creating List containing Unique Genre
unique_Genre = []
for i in one_D:
    if i not in unique_Genre:
        unique_Genre.append(i)
```

```python
unique_Genre
```
